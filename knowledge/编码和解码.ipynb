```python
import sys

# 创建一个包含一个汉字字符的字符串
chinese_character = '中'

# 获取字符串的字符数
character_count = len(chinese_character)

# 获取字符串对象占用的内存字节数
memory_size = sys.getsizeof(chinese_character)

print(f'字符数: {character_count}')
print(f'内存字节数: {memory_size} 字节')
```

在 Python 中，decode 方法主要用于将字节序列解码为字符串，通常用于处理编码的数据。下面是一个使用 decode 方法的示例：

```python
# 创建一个字节序列（bytes）
encoded_bytes = b'\xe4\xb8\xad\xe6\x96\x87'  # 这是 UTF-8 编码的 "中文" 字符串

# 使用 decode 方法将字节序列解码为字符串
decoded_string = encoded_bytes.decode('utf-8')

# 打印解码后的字符串
print(decoded_string)
```

In the Python programming language, encoding represents a Unicode string as a string of bytes. This commonly occurs when you transfer an instance over a network or save it to a disk file. Decoding transforms a string of bytes into a Unicode string. This happens when you receive a string of bytes from a disk file or the network.
